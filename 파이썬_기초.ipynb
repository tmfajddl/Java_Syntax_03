```python
#딕셔너리: 시퀀스가 아닌 자료형 / 순서가 없는 자료형
#데이터의 저장방식이 특이함: {}안에 한쌍의 데이터로 저장해야한다.
#{키:밸류} 의 한쌍으로 저장
#키: 내가 저장해둔 데이터를 사용할 떄 해당 데이터를 찾아갈 수 있는 근거
#밸류: 내가 저장해둔 데이터

a = {'키1': '밸류1', '키2' : '밸류2'}
print(a['키1'])

a = {'키1': '밸류1', '키2' : '밸류2'}
a['키1'] = '수정된 밸류1' #데이터 수정
a['키3'] = '밸류3' #데이터 추가
a['키2'] = '밸류2??' #데이터 수정
#키는 고유값으로 절대로 중복될 수 없다.
print(a)

a = {'키1': '밸류1', '키2' : '밸류2'}
del a['키1'] #데이터 지움
print(a)

a = {'키1': '밸류1', '키2' : '밸류2'}
print(a.keys()) #keys: 키들을 리스트로 변환하여 dict_keys type으로 변환

a = {'키1': '밸류1', '키2' : '밸류2'}
print(a.values()) #values: 밸류들을 리스트로 변환하여 dict_values type으로 변환

a = {'키1': '밸류1', '키2' : '밸류2'}
print(a.items()) #items: 키와 밸류들을 튜플로 만든 후 리스트로 변환하여 dict_items type으로 변환

#키는 문자든 숫자든 상관 없음 단, 가변 자료형은 안댐(리스트, 딕셔너리)/ 같은 딕셔너리가 아니라면 키 이름이 같아도 중첩이 아님
#딕셔너리는 한줄로 쓰지 않음
a = {
    '1' : {
        '1' : 10,
        '2' : 20 #1
    },
    '2' : 20 #2
}
print(a['2']) #2의 20 출력
print(a['1']['2']) #1의 20 출력
print('2' in a) #True / a=a.keys()로 보면 댐
print(20 in a) #Felse / 키로만 검색이 된다.

a = {'키1': '밸류1', '키2' : '밸류2'}
#print(a['키3']) #에러로 뜸
print(a.get('키3')) # 키3이 없어도 에러가 안뜸 None출력 / 그래서 a.get을 주로 사용해야한다
print(a.get('키3', '키3은 존재하지 않습니다')) # '키3은 존재하지 않습니다'출력 / 메세지를 추가할 수 있음

a = {'키1': 2, '키2' : 5}
a['키3'] = a.get('키3', 0) + 1 #만약 a.get('키3', 0) 자리에 a['키3']을 넣으면 출력 안댐
print(a)

scores = {
    '수학' : 80,
    '과학' : 95,
    '영어' : 100
}
highScores = {k : v for k, v in scores.items() if v >= 85}
#highScores = {}
#for k, v in scores.items():
# if v >= 85:
#   highScores(k) = v
print(highScores)

a = {
    1 : 10,
    2 : 20
}
b = {
    1 : 30,
    3 : 40
}

print(a|b) #| : 앞에 있는거에 뒤에 있는거를 합치다 = a.update(b)

```

```python
#set(집합): {}를 사용하나 딕셔너리와 달리 값만 존재함
#순서가 의미가 없음 = 인덱싱이 안됨 / 슬라이싱 불가
#값의 중복을 허용하지 않는다. / 딕셔너리는 키만 중복을 허용하지 않았지만 값의 중복은 가능했음 -> 가장 큰 차이점
#무수히 많은 데이터 중 중복을 제거할 때 사용 / 각 데이터 묶음끼리 합집합, 교집합, 차집합 등의 개념을 적용시킬 때 사용
#가변 데이터이다.
s = {1,2,2,3,4}

print(s) # {1,2,3,4} 출력 / 중복된 것을 제거해버림

s1 = set([1,2,2,3,4]) #리스트로 만든 후에 set으로 변환해줘야함
print(s1)

s1 = set ('hello')
print(s1) #{'e', 'o', 'l', 'h'} 츌력 : 순서에 아무 상관이 없다. / 인덱스가 부여되지 않는다.

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(s1 & s2) # &: 교집합 출력
print(s2.intersection(s1)) # .intersection : 교집합 출력

print(s1|s2) # |: 합집합 출력
print(s1.union(s2)) # .union : 합집합 출력

print(s1-s2) # -: 차집합 출력
print(s1.difference(s2)) # .difference : 차집합 출력

s1.add(6) # 값을 하나씩만 추가할 수 있음
s1.add('7') # 숫자와 문자를 동시에 저장 가능
print(s1)

s2.update([9,10]) # 값 여러개를 리스트 형태로 넣어서 추가할 수 있음
print(s2)

# 만약 집합 안에 요소 중 하나를 꺼내고 싶다면?
# 1. 반복문으로 돌린다.
# 2. 리스트나 튜플로 만든다.

s1.remove(3)
s1.add(30)
print(s1) # 값의 변환은 불가 = 지우고 다시 추가하는 방향으로 진행해야 함

s1.remove(2) # remove를 하나의 값을 연속적으로 지운다하면 에러뜸
s1.discard(2) # discard는 하나의 값을 연속적으로 지워도 에러 안뜸
print(s1)

# 가변 데이터의 얕은 복사: 객체의 효율적인 사용을 위해 존재 / 리스트, 딕셔너리, 집합에만 적용
s1 = {1,2,3}
s2 = s1
s2.add(4)

print(s1) #{1, 2, 3, 4}출력 / 리스트는 공유하고 s1,s2라는 리모컨을 공유한다고 생각
print(s2) #{1, 2, 3, 4}출력

# 얕은 복사2 (.copy = [:])
s1 = [[1,2],[2,3]]
s2 = s1.copy() # .copy : 새로운 리스트를 만들지만 가장 바깥쪽 리스트만 해당 됨 / 안쪽 리스트는 얕은 복사
s2[0][0] = 100

print(s1)
print(s2)

# 깊은 복사: 불변 자료형들처럼 각각의 독립적인 데이터로 처리하는 방법
import copy # 이름 명시

s1 = {1,2,3}
s2 = copy.deepcopy(s1) # .deepcopy : 깊은 복사를 하는 메서드
s2.add(4)

print(s1) #{1, 2, 3}출력 / s2에 요소를 추가했지만 s1에는 영향이 없음
print(s2) #{1, 2, 3, 4}출력

#만약 []가 하나라면? .copy나 [:]는 깊은 복사다
s1 = [1,2,3]
s2 = s1.copy()
s2[0] = 100

print(s1)
print(s2)
```

```python
#매서드
#인스턴스 메서드(첫번째 매개변수로 self가 있어야한다)
class Person:
  name = '홍길동'
  def hello(self):
    print(Person.name)

p = Person()
p.hello()

#클래스 메서드: 클래스의 상태를 관리하기위해 사용하는 메서드
class Person:
  test = 'test'
  def __init__(self):
    self.test2 = 'test2'
  @classmethod #데코레이터: 클래스 메서드임을 명시
  def test3(cls):
    print('class method 실행')
  def test4(self):
    print('instance method 실행')

a = Person()
a.test3() #클래스 메서드는 인스턴스로 호출 가능 / 하지만 클래스를 이용하여 호출하는개 더 적합함
a.test4() #인스턴스 메서드는 인스턴스로 호출 가능
Person.test3() #클래스 메서드는 클래스도 호출 가능
#Person.test4() : 인스턴스 함수는 클래스로 호출이 불가하다. why? 객체를 만들지 않았기 때문에 인자가 없다.
Person.test4(Person()) # 만약 인자를 넣어줄 경우 실행 가능

#1. 클레스 메서드는 인스턴트 속성을 사용할 수 없다
class Person:
  test = 'test'
  def __init__(self):
    self.test2 = 'test2'
  @classmethod
  def test3(cls):
    print('class mathod 실행')
    print(Person.test) # = print(cls.test)
    # print(self.test2) #실행안댐
    cls.test4(Person()) # 이렇게 인자를 넣어주면 인스턴스 호출 가능 / 이때 인자는 객체 여야함 (cls 안댐)
  def test4(self):
    print('instance mathod 실행')
    print(Person.test)
    print(self.test2) # test3을 통해 test4로 넘어오고 그 뒤 test2 실행은 가능

Person.test3()
Person.test4(Person())

#스태틱 메서드: 인스턴스 속성을 사용할 수 없음 / 클래스 속성은 사용할 수 있음
#클래스 메서드와 차이점: 매개변수로 cls를 가지지 않는다->class나 instance 요소에 관여 안함
#유티릴티 메서드를 클래스에 종속시키기위해 사용

class Person:
  test = 'test'
  def __init__(self):
    self.test2 = 'test2'
  @staticmethod
  def test3():
    print('stattic method 실행')

a = Person()
a.test3()
Person.test3() #클래스로 호출하는 것을 선호
```

```python
class Parent:
  name = '부모1'

  @staticmethod
  def changeName(newName):
    Parent.name = newName

# class의 상속: super class가 가진 속성을 모두 물려줌
class Child(Parent): # Parent라는 super class가 가진 속성을 상속 시켜줌
  pass

p = Parent()
c = Child()

p.changeName('부모2')
print(Parent.name, c.name)

c.changeName('자식1')
print(Parent.name, c.name)
#스테틱 메서드는 슈퍼 클래스나 서브 클래스 하나만 바꿔도 둘다 바뀜

class Parent:
  name = '부모1'

  @classmethod
  def changeName(cls,newName):
    cls.name = newName

class Child(Parent):
  pass

p = Parent()
c = Child()

p.changeName('부모2')
print(Parent.name, c.name)

c.changeName('자식1')
print(Parent.name, c.name)
#클래스 메서드는 슈퍼클래스는 서브클래스에 영향을 주나 서브클래스는 슈퍼클래스에 영향을 주지 못함
#why? cls때문에 무조건 Parent class로 넘어감
```

```python
# Employee 클래스와 Department 클래스를 작성하세요. 요구 사항은 다음과 같습니다.

# Employee 클래스
# name과 position을 인스턴스 속성으로 가집니다.
# 생성자를 통해 이름과 직책을 초기화합니다.

# Department 클래스
# name과 직원 목록(employees)을 인스턴스 속성으로 가집니다.
# 생성자를 통해 부서 이름을 초기화하고, 직원 목록은 빈 리스트로 초기화합니다.
# add_employee 메서드를 작성하여 Employee 객체를 추가합니다.
# 부서에 속한 직원들의 이름을 출력하는 list_employees 메서드를 작성하세요.
# Department 객체를 생성하고 여러 Employee 객체를 추가한 뒤, 직원 이름 목록을 출력하세요.
class Employee:
  def __init__(self,name,position):
    self.name = name
    self.position = position

class Department:
  def __init__(self,name):
    self.name = name
    self.employees = []
  def add_employee(self,emp):
    self.employees.append(emp)
  def list_employees(self):
    for v in self.employees:
     print(f'직원이름 : {v.name}, 직책 : {v.position}')
dept = Department('기술개발부')
emp1 = Employee('신짱아','차장')
emp2 = Employee('신짱구','부장')
emp3 = Employee('김철수','대리')

dept.add_employee(emp1)
dept.add_employee(emp2)
dept.add_employee(emp3)

dept.list_employees()
```

```python
# Product 클래스를 작성하세요. 요구 사항은 다음과 같습니다.

# name, price, stock을 인스턴스 속성으로 가집니다.
# 생성자를 통해 이름, 가격, 재고를 초기화합니다.
# 가격은 0보다 커야 하며, 그렇지 않으면 ValueError를 발생시킵니다.
# 재고는 음수일 수 없으며, 음수 값이 들어오면 0으로 초기화합니다.
# sell 메서드를 작성하여 재고를 감소시키고, 재고가 부족하면 경고 메시지를 출력합니다.
# Product 객체를 생성하고 메서드를 호출하여 동작을 확인하세요.

class Product:
  def __init__(self,name,price,stock):
    self.name = name
    if price > 0:
     self.price =price
    else:
      raise ValueError('가격은 0보다 커야합니다')
    if stock >= 0:
      self.stock = stock
    else:
      self.stock = 0
  def sell(self,quantity):
    if quantity >self.stock:
      print(f'{quantity}개를 판매할 수 없습니다. 현재 재고는 {self.stock}개입니다')
    else:
      self.stock -= quantity
      print(f'{self.name}을(를) {quantity}개를 판매하였습니다. 현재 재고는 {self.stock}개입니다')

try: #try: 예외처리 / try를 실행 중 문제가 발생 된다면 except로 넘어감
#개발자가 오류가 일어날 것을 예측하여 예외처리 해줌으로써 프로그램이 멈추는것을 방지함
 prod = Product('커피',2000,10)

 prod.sell(6)
 prod.sell(5)


except ValueError as e: # as e: 메세지가 가지고 있는걸 보여주기 위해 넣은 것
  print(e)
```
